```python
from sympy import symbols, diff, solve, limit, oo

# Define the symbolic variables
p, N = symbols('p N', real=True, positive=True)

# Define the efficiency function of slotted ALOHA
E = N * p * (1 - p)**(N - 1)

# Calculate the derivative of E with respect to p
dE_dp = diff(E, p)

# Find the value of p that maximizes E for each N
max_efficiency_points = []
for N_value in range(1, 101):  # Sweep over N from 1 to 100
    # Solve for the critical points of E
    critical_points = solve(dE_dp.subs(N, N_value), p)
    
    # Since we know that 0 < p < 1, we filter out any solutions outside this range
    critical_points = [point.evalf() for point in critical_points if 0 < point.evalf() < 1]
    
    # Append the results to the list
    max_efficiency_points.append((N_value, critical_points))

# Print out the results
for N_value, points in max_efficiency_points:
    print(f'N = {N_value}: p = {points}')

# Bonus: Find the limit of E/N as N approaches infinity, knowing that optimal p is around 1/N
E_over_N = E / N
limit_expression = limit(E_over_N.subs(p, 1.0/N), N, oo)

print(f'The limit of E/N as N approaches infinity is {limit_expression}')
```
